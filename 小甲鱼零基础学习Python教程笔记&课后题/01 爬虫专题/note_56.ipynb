```python
#download_pic1.py
import urllib.request
import os 

def download_pic(folder='pic'):
    os.mkdir(folder)
    os.chdir(folder)
    url = 'http://jandan.net/top-ooxx'
    request = urllib.request.Request(url)
    responce = urllib.request.urlopen(request)
    html = responce.read().decode('utf-8')
    print(html)
    
download_pic()
    
    
    
    



```
