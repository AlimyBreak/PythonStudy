# 一、修改headers的字段 User-Agent
## 1、Request对象生成时，有header输入参数可以传入
## 2、Request对象生成后，通过Request对象的add_header()方法来设置

```python
import urllib.request
header = {}
data = {};
header['User-Agent'] = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0'
url='http://www.baidu.com'
#request = urllib.request.Request(url,data,header);
request = urllib.request.Request(url,data)
request.add_header('User-Agent','Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0')
responce = urllib.request.urlopen(request)
html = responce.read().decode('utf-8')
print(html)
```

# 二、代理
## 1、代理是什么?
###  代理帮你访问，访问到的内容转发给原始请求的用户
## 2、代理步骤
### a.参数是一个字典{'类型':'代理ip:端口号'}
###    proxy_support = urllib.request.ProxyHandler({})
### b.定制、创建一个opener
###    opener = urllib.request.build_opener(proxy_support)
### c.安装一个opener 影响全局
###   urllib.request.install_opener(opener)
### d.直接使用opener 不影响全局
###   opener.open(url)

```python
import urllib.request
url = 'http://www.ip138.com'
#url = 'http://www.whatismyip.com.tw'
proxy_support = urllib.request.ProxyHandler({'http':"118.180.166.195:8060"})
opener = urllib.request.build_opener(proxy_support)
opener.addheaders = [('User-Agent','Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0')]
urllib.request.install_opener(opener)
responce = urllib.request.urlopen(url)
html = responce.read().decode('gb2312','ignore')
print(html)
#貌似没有返回 "您的IP是：[xxx.xxx.xxx.xxx] 来自：xxxx"

```
